# Error handling

In order to handle errors in a proper way we can use the try: except: syntax. All the code that we get into the try is everything that the program will try to execute and if there is an error we will use the except to work with that error.

We can specify the errors that we want to handle in the except statement or leave it blank so that any error will go to that statement.

````python
def handlingErrors(value):
    try:
        num = int(value)
        print('The number you typed is: {}'.format(num))
    except ValueError:#ValueError is a type of error, it's an optional entry for the try/except statement
        print("You didn't typed a number, you typed: ", value)

handlingErrors(1)
handlingErrors('Hey, this is an error')
````

# Fundamentals of OOP

## Encapsulation

Define the scope for some attributes. It's useful to reduce memory usage and extra executions.

We can define attributes that will only be accessible within the class and not outside of it.

In order to encapsulate an attribute when we are defining it we have to write them as self.__privateAttribute. Then in order to access it (write/read) we are only going to be able to access it within the class using getters and setters methods. It's a best practice when coding with OOP.

Another way to define private attributes is by self._privateAttribute, the difference between this syntax and the previous one is that this is a best practice between programmers but it won't tell python that the attribute must be private. Is the most common standard.

### Get method (Getter)

Access the value of attributes using get methods. The method should be called as the attribute. As it is a method we should use it using () but this is not a friendly feature for the programmers, in order to call a method as it was an attribute, without the (), we just have to add the @property keyword above the method.

There must be a getter for each attribute, is a best practice.

### Set method (Setter)

Each attribute could change it's value during the lifecycle of an object, in order to do so with the best practices we have to use the setters. 

In order to create a setter we have to specify the decoration of the method which is @attribute.setter and then the function.

```python
class myClass():
    def __init__(self):
        self._attribute = 1

    @property
    def attribute():
        return self_attribute

    @attribute.setter
    def attribute(self, attribute):
        self._attribute = attribute

objectInstance = myClass()

print(objectInstance.attribute)#Using the getter of the class for that attribute
objectInstance.attribute = 2#Using the setter of the class for that attribute
```

If we only create a getter but not a setter for an attribute it will be created as read-only. We must create the getter of an attribute.



# Inheritance

Between classes, all the methods from the parent class will be automatically created in the child class.

In order to indicate an inheritance between classes we just have to create the child class and inside the () indicate the class from which it will inherit.

```python
class Person():
    def __init__(self, name, age):
        self._name = name
        self._age = age
    
    def sayHi(self):
        print("Hey, I am {} and I'm {} years old".format(self._name, self._age))

class Doctor(Person):#The Doctor class will inherit from Person
    pass
```

While working with inheritance, if we add a new attribute to the child class from it's own init method and we want to inherit all the attributes that were created in the parent class, we must add the supper keyword to the child class. If the child class has an attribute or method with the same name and parameters as the parent class, python will not inherit that component, instead it will use the one we created in the child class.

## Multiple inheritance

We can inherit from multiple classes with python, in order to do so, while creating a child class we just have to add al the classes from the child class will inherit separated by commas.

# Polymorphism

Its another fundamental concept from OOP. It refers to when we have to modify a method that the child class inherited from it's parent class or when we execute the same code with different objects.

```python
class Animals():
    def __init__(self, age, name):
        self._age = age
        self._name = name

    def walk(self):
        print("I'm a {} and I can walk!".format(self._name))

class Dog(Animals):
    def walk(self):
        print('I can walk with 4 legs')

class Kangaroo(Animals):
    def walk(self):
        print('I can walk with 2 legs')

myDog = Dog(5, 'Koda')
myDog.walk()
myKangaroo = Kangaroo(10, 'Jack')
myKangaroo.walk()
```
