## Creating substrings



With python there are multiple ways for creating a substring, they are zero based.

Here are some examples:

```python
#Operators with strings

string = 'Test'

print(string * 3, '\n')

substrings = 'This is a substring example'
print('original string: ', substrings)
print('a single char: ', substrings[1], '\n')
print('From : until (without including this index: )',substrings[6:10], '\n')
print('From the first position (index= 0) : until (without including this index: )',substrings[:10], '\n')
print('From : until the last character of the string: )',substrings[10:], '\n')
print('Negative indexes, they will start backwards: )',substrings[:-5], '\n')
#Creates a copy of the string (first :) the second : are for traverse the complete string and the number represents the steps/jumps in which will store the characters in this example it will write every two characters
print("Print all even index characters: " + substrings[: : 2])
#To print a string backwards
print("To print a string backwards: " + substrings[: : -1])
```
